```python
X = rand.rand(10,2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
plt.scatter(X[:, 0], X[:,1], s=100);
```

```python
differences = X[:, np.newaxis, :] - X[np.newaxis, :, :]
differences.shape
```

```python
sq_differences = differences ** 2
sq_differences.shape
```

```python
dist_sq = sq_differences.sum(-1)  # 마지막 축 axis=2에 해당
dist_sq.shape
```

```python
dist_sq.diagonal()
```

```python
nearest = np.argsort(dist_sq, axis=1)
print(nearest)
```

```python
nearest
```

```python
K = 2
nearest_partition = np.argpartition(dist_sq, K + 1, axis=1)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=100)
K = 2
for i in range(X.shape[0]):
    for j in nearest_partition[i, :K+1]:  # i에 가장 가까운 k+1개
        plt.plot(*zip(X[j], X[i]), color='black')  # zip()은 연결선 처리
```
